```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
# using a fixed seed for reproducibility
numpy.random.seed(7)
```

binary classification problem, meaning 0 or 1. in this dataset, people either are on the road to having diabetes or are safe from diabetes

```python
# load pima indians dataset
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

dataset has 9 columns and the range 0:8 will select columns from 0 to 7

## Defining the model

Ensure the input layer has the correct number of inputs. This can be done by specifying the "input_dim" argument when creating the layer.

In this example, we create a fully-connected network structure with three layers.
Fully connected networks are defined using the Dense class.

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation="relu"))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

## Compiling the model

Compilation step uses Tensorflow under the hood. Backend automatically decides if it should use the CPU, GPU or a mix of both.

When compiling, we must specify additional properties required when training the network.

**NOTE**: Training a network means finding the best set of weights to make predictions for the problem

```python
# compile model
model.compile(loss="binary_crossentropy", optimizer='adam', metrics=['accuracy'])
```

## Fit model - training

we can train or fit our model on our loaded data by calling the fit() function on the model.


The training process will run for a fixed number of iterations through the dataset called epochs. We must specify the number of epochs using the "epochs" argument.

```python
# Fit the model
model.fit(X, Y, epochs=150, batch_size=10)
```

## Evaluating the model

ideally, we could have separated our datasets into training and test datasets. Test datasets would be used to evaluate our model's accuracy and loss

```python
# evaluates the model
scores = model.evaluate(X,Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
print(model.metrics_names)
```

## Making predictions

making predictions with keras can be done with model.predict(). Since we are using the sigmoid activation function on the output layer, the predictions will be in the range between 0 and 1. We can convert the predictions into binary predictions by rounding the outputs.

```python
# calculate predictions
predictions = model.predict(X)
```

```python
print(predictions)
```

```python
rounded = [round(x[0]) for x in predictions]
```

```python
print(rounded)
```
